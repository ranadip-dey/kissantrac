```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import urllib.request as urllib2
import itertools
import numpy as np

        
def make_soup(url):
    req = urllib2.Request(url, headers={'User-Agent' : "Magic Browser"}) 
    html = urllib2.urlopen(req)
    return BeautifulSoup(html, 'html.parser')


def get_data(url):
    soup = make_soup(url)
    page_body = soup.body
    first_h1 = soup.select('h1')[0].text
    filename= "khetigaadi_" + first_h1.replace(" ","_") + ".csv"
    print( filename )
    #get drive
    all_attr_header = []
    all_attr_val = []
    all_desc = []
    all_feature = []
    all_spec = []
    work_val = ""

    all_drive = page_body.find_all('span', attrs={'class':'tractor-drive'})
    for drive in all_drive:
        #print(drive.text)
        all_attr_header.append("Wheel Drive ")
        all_attr_val.append(drive.text)
        
    table = soup.find("table", "table main-overview-panel__key-points-list")
    for row in table.findAll("tr"):
        colHs = row.findAll("td")
        for ch in colHs:
            all_attr_header.append(ch.text)
        cells = row.findAll("th")
        for col in cells:
            all_attr_val.append(col.text)

    tableBelow = page_body.find_all("section", attrs={'section-wrap tab-info-section'})
    if len(tableBelow)==2:
        for vl in tableBelow:
            work_val=vl
#        print(vl)
        table2 = work_val.find("table")
        if table2:
            desc= work_val.find_all("p")
            for d in desc:
                #print("Description : " + str(d))
                all_desc.append(d.text)
            for tag in work_val.find_all("li"):
                all_feature.append(str(tag.text).replace("<li>","").replace("</li>",""))
            for ro in table2.find_all("tr"):
                row = ro.findAll("td")
                if len(row)==1:
                    csv_row = row[0].get_text()
                    all_spec.append(csv_row)
                if len(row)==2:
                    csv_row = row[0].get_text().replace("\n","") + "," + row[1].get_text().replace("\n","")
                    all_spec.append(csv_row)
                    
     #write in file
    with open(filename , 'w') as output_file:
        output_file.write(first_h1)
        output_file.write("\n")
        output_file.write("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
        output_file.write("\n")
        for item in list(zip(all_attr_header, all_attr_val)):
            output_file.write(str(item).replace("(","").replace(")","").replace(":","").replace("'","") + "\n")
        output_file.write("\n")
        output_file.write("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
        output_file.write("\n")
        output_file.write("\n")
        for ds in all_desc:
            output_file.write(ds.replace(",",""))
        output_file.write("\n")
        output_file.write("Features")
        output_file.write("\n")
        for ft in all_feature:
            output_file.write(ft)
            output_file.write("\n")            
        output_file.write("\n")
        for sp in all_spec:
            output_file.write(str(sp) + "\n")
      


#get_data("https://khetigaadi.com/new-tractor-model/mahindra%C2%A0jivo%C2%A0225%C2%A0di%C2%A04wd/en")
# get_data("https://khetigaadi.com/new-tractor-model/farmtrac-champion-42-valuemaxx/en")
cn=1
file1 = open('khetigaadi_Eicher.csv', 'r')
Lines = file1.readlines()

for line in Lines:
    line= line.rstrip()
    print(cn)
    get_data(line)
    cn+=1
print("all file downloaded.")
```
